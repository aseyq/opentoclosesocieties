```python
from commons.agent import Agent
from commons.couple import Couple

# Create two agents
m1 = Agent(gender="male")
f1 = Agent(gender="female")
print("Agent 1", m1)
print("Agent 2", f1)

# Create a couple
couple1 = Couple(f1, m1)
print("Couple 1", couple1)

print(m1.name, "'s spouse is", m1.spouse.name)
print(f1.name, "'s spouse is", f1.spouse.name)
```

```python

# Lets put them in a community
from commons.community import Community
community = Community()

print("Community", community)
# By default, we have an egalitarian community with a random wealth.

community.add_agent(m1)
community.add_agent(f1)

print("Community after adding agents", community)

print("====Agents in the community====")
for i in community.agents:
    print("- ", i)
```
